```python
import os
import glob
import pandas as pd
import xml.etree.ElementTree as ET
import csv
```

```python
project_folder_path='/home/hemanth/Desktop/Multi_OD_Pytorch/'
```

```python
def xml_to_csv(path,folder):
    xml_list = []
    for xml_file in glob.glob(path + '/*.xml'):
        tree = ET.parse(xml_file)
        root = tree.getroot()
        for member in root.findall('object'):
            value = (project_folder_path+'dataset/'+ folder+'/' +root.find('filename').text,
                     int(member[4][0].text),
                     int(member[4][1].text),
                     int(member[4][2].text),
                     int(member[4][3].text),
                     member[0].text
                     )
            print(value)
            xml_list.append(value)
    xml_df = pd.DataFrame(xml_list)
    return xml_df
```

```python
for folder in ['train','test']:
    image_path = os.path.join(os.getcwd(), (project_folder_path+'dataset/' + folder))
    xml_df = xml_to_csv(image_path,folder)
    if not os.path.exists(project_folder_path+'dataset/' + folder + '_labels.csv'):
        open(project_folder_path+'dataset/' + folder + '_labels.csv','w')
    xml_df.to_csv((project_folder_path+'dataset/' + folder + '_labels.csv'), index=None)
    print('Successfully converted xml to csv for dataset : ',folder, '\n\n')
```
